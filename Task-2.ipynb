```python
#Active id in the whole dataset
a = pd.DataFrame(data['Id'].value_counts())
a = a.reset_index()
for i in range(len(a["Id"])):
    if(max(a['Id'])==(int(a[i:i+1]['Id']))):
        print("Active person is  : ", int(a[i:i+1]['index']))
```

```python
#how many of them not logged

len(set(data["Id"])) - len(set(data[data["LoggedActivitiesDistance"]>0].Id))
```

```python
#id have the laziest id in the whole dataset

b = pd.DataFrame(data["Id"].value_counts())

b = b.reset_index()

for i in range(len(b["Id"])):
    if(max(b["Id"])==(int(b[i:i+1]["Id"]))):
        print("Most laziest id : " , int(b[i:i+1]["index"]))
```

```python
# calories burn required for the healthy person

# for the daily routine
# female required (1,600 – 2,200 ) per day
# male required (2,200 – 3,000) per day

calories =  len(set(data[data["Calories"]>2400].Id))
calories
```

```python
# how many person is not regualr per day , for the normal person it wold take 8000 steps per day and data is around for 21 days

a = pd.DataFrame(data.groupby('Id')['TotalSteps'].sum())
a.reset_index()
a.head(3)
```

```python
normal_person = len(a[a['TotalSteps']>168000].index.to_list())
print("Total normal person  : " , normal_person)
```

```python
not_normal_person = len(a[a['TotalSteps']<168000].index.to_list())
print("Total normal person  : " , not_normal_person)
```
